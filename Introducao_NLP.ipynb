### 1.2 Tokenização

O processo de tokenização tem como objetivo dividir palavras em tokens, ou seja, unidades de texto.  Podem ser palavras ou pontuação.

```python
#importando as bibliotecas especificas para tokenização
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
#Livro Harry Potter e a Ordem da Fênix, página 274
exemplo = "Harry aguardava com ansiedade a viagem de fim de semana a Hogsmeade, mas uma coisa o preocupava. O Sirius mantivera silêncio absoluto desde que aparecera no fogo no começo de Setembro;"
```

```python
print(exemplo)
```

```python
#divisão em sentenças
sent_tokenize(exemplo)
```

```python
#armazendo em uma variável a divisão por palavras
ex_tokenized = word_tokenize(exemplo)
```

```python
print(ex_tokenized)
```

### 1.3 Stopwords

Tratam-se de palavras que não interferem no contexto de análise, podendo ser excluídas durante o processo.

```python
#importar a biblioteca que contem as stopwords
from nltk import corpus

#baixar a lista de stopwords
nltk.download('stopwords')

#armazenar em uma variável a lista de stopwords
stopwords = corpus.stopwords.words('portuguese')
```

```python
#lista de stopwords
print(stopwords)
```

```python
#número de tokens no exemplo

len(ex_tokenized)
```

```python
#Montando uma lista com o exemplo sem as stopwords

exemplo_semstopwords = []

for i in ex_tokenized:
  if i not in stopwords:
    exemplo_semstopwords.append(i)
  else:
    pass
```

```python
print(exemplo_semstopwords)
```

### 1.4 Lemmatization 

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
#instanciando o lemmatizador
lemmatizer = WordNetLemmatizer()

print('preocupava:', lemmatizer.lemmatize('preocupava'))
print('mantivera:', lemmatizer.lemmatize('mantivera'))
```

```python
exemplo2 = "Harry was looking forward to the weekend trip to Hogsmeade, but one thing worried him. Sirius had kept absolute silence since he appeared at the fire in early September;"
```

```python
#Tokenização
ex2_tokenized = word_tokenize(exemplo2)

#Stopwords
stopwords2 = corpus.stopwords.words('english')

exemplo2_semstopwords = []

for i in ex2_tokenized:
  if i not in stopwords2:
    exemplo2_semstopwords.append(i)
  else:
    pass

#Lemmatization
print('worried:', lemmatizer.lemmatize('worried', pos = 'v'))  #estamos especificando que é um verbo
print('looking:', lemmatizer.lemmatize('looking', pos = 'v'))
```

### 1.5 Stemming

```python
from nltk.stem import PorterStemmer
```

```python
#instanciando o stemmer
stemmer = PorterStemmer()

print('preocupava:', stemmer.stem('preocupava'))
print('mantivera:', stemmer.stem('mantivera'))
```

```python
print('looking:', stemmer.stem('looking'))
print('worried:', stemmer.stem('worried'))
```

### 1.6 Bag-of-words

```python
!pip install sklearn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#lista de textos para a análise
texto = ["Assisti um filme ótimo","Assisti um filme ruim"]

#armazena na variável vetorizer a função para gerar a Bag-of-words
vetorizer = CountVectorizer(lowercase=False)

#transformando o texto em Bag-of-words
bag_of_words = vetorizer.fit_transform(texto)
```

```python
vetorizer.get_feature_names()
```

```python
bag_of_words
```

Para melhor visualizar a matriz da bag-of-words, transformaremos a mesma em matriz esparsa utilizando uma função da biblioteca pandas:

```python
matriz_esparsa = pd.DataFrame.sparse.from_spmatrix(bag_of_words,
                     columns=vetorizer.get_feature_names())
```

```python
matriz_esparsa
```

### 1.7 WordCloud

```python
#importando as bibliotecas matplotlib e wordcloud
%matplotlib inline
import matplotlib.pyplot as plt

from wordcloud import WordCloud
```

Utilizaremos como base a definição de nuvem de palavras no Wikipedia.

```python
definicao = [
             "Nuvem de tags, nuvem de palavras ou nuvem de etiquetas é uma lista hierarquizada visualmente, uma forma de apresentar os itens de conteúdo de um website.",
             "Tag (tég) é 'etiqueta', 'rótulo' ou 'legenda'. Pense na etiqueta como uma categoria simples. As pessoas em geral podem categorizar páginas web, fotos, e vídeos com qualquer etiqueta que faça sentido. O uso das etiquetas na web foi difundido por sites como Technorati, del.icio.us e Flickr, e o aproveitamento inteligente das novas possibilidades de classificação em interfaces inovadoras significou um alento aos internautas ávidos por melhor sinalização para navegação no mar de conteúdo da rede.",
             "Uma nuvem de etiquetas em geral reúne um conjunto de etiquetas utilizadas em um determinado website disposto em ordem alfabética, e a quantidade de conteúdos que o site apresenta em cada etiqueta é mostrado proporcionalmente pelo tamanho da fonte. Dessa forma, em uma mesma interface é possível localizar uma determinada etiqueta tanto pela ordem alfabética como pela frequência da incidência de conteúdos marcados com a mesma etiqueta no referido site. As etiquetas disponibilizadas na nuvem são links que levam a coleções de itens relacionados às palavras da etiqueta."
]
```

```python
#unindo todas as palavras da lista definicao
todas_palavras = ' '.join([texto for texto in definicao])
```

```python
#criando a nuvem de palavras
nuvem_palavras = WordCloud(width = 800, height = 500, max_font_size = 110,
                          collocations=False).generate(todas_palavras)
```

```python
#criando a imagem de nuvem de palavras
plt.figure(figsize = (10,7))
plt.imshow(nuvem_palavras, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
